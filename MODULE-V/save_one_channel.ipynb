```
import cv2
import numpy as np
img = cv2.imread('cat.jpg')
```

```
print(img.shape)
```

```
b, g, r = cv2.split(img)
b = img[...,0]
g = img[...,1]
r = img[...,2]
```

```
print(b.shape)
```

```
import matplotlib.pyplot as plt
plt.imshow(b, cmap='Blues')
plt.show()
plt.imshow(g, cmap='Greens')
plt.show()
plt.imshow(r, cmap='Reds')
plt.show()
```

```
print(b[200:210, 200:210])
```

```
np.savetxt('one-channel.csv', b, fmt='%.3f', delimiter=',')
```
